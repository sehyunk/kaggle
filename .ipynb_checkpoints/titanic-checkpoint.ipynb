# 전처리
## Age 결측치 처리

```python
for df in dfs:
    for i in list(df['Initial'].value_counts().index):
        mean = df.loc[df['Initial'] == i]['Age'].mean()
        print(f'{i}: {round(mean)}')
        df.loc[(df['Age'].isnull()) & (df['Initial'] == f'{i}'),'Age'] = mean
train['Age'].isna().sum()
```

## Age 카테고리화

```python
for df in dfs:
    print('---------')
    for i in range(4,0,-1):
        globals()[f'q{i}'] = df['Age'].quantile(q = i/4)
        print(f'q{i}:  ~ ' + str(round(globals()[f'q{i}'])))
        for j in df.index:
            if df.loc[j, 'Age'] <= globals()[f'q{i}']:
                df.loc[j, 'Age_q'] = f'q{i}'

sns.countplot(train['Age_q'])
```

## one hot encoding

```python
train = pd.get_dummies(train, columns = ['Pclass', 'Sex', 'Embarked', 'Initial', 'Age_q'])
test = pd.get_dummies(test, columns = ['Pclass', 'Sex', 'Embarked', 'Initial', 'Age_q'])
```

```python
train.head(3)
```

# Modeling

```python
X = train.iloc[:, 1:]
y = train.iloc[:,0]
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kfold = KFold(n_splits = 10, shuffle = True, random_state = 0)
from sklearn.model_selection import GridSearchCV
```

## knn classifier

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
param_grid = {'n_neighbors': list(range(1,30))}
scores = cross_val_score(GridSearchCV(clf, param_grid, cv = kfold),
                         X_train, y_train, cv = kfold)
print('''
scores: {} 
 - - - - - - - - - -
ㅣ score_mean: {:.2f} ㅣ
 - - - - - - - - - -
'''.format(scores, scores.mean()*100))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
param_grid = {'C':[0.0001,0.001,0.01,0.1,1,10,100,1000,10000]}
scores = cross_val_score(GridSearchCV(clf, param_grid, cv=kfold),
                         X_train, y_train, cv=kfold)
print('''
scores: {} 
 - - - - - - - - - -
ㅣ score_mean: {:.2f} ㅣ
 - - - - - - - - - -
'''.format(scores, scores.mean()*100))
```

## SVM

```python
from sklearn.svm import SVC
clf = SVC(C = 10, gamma = .1)
scores = cross_val_score(clf, X_train, y_train, cv = kfold)
print('''
scores: {} 
 - - - - - - - - - -
ㅣ score_mean: {:.2f} ㅣ
 - - - - - - - - - -
'''.format(scores, scores.mean()*100))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
scores = cross_val_score(clf, X, y, cv=kfold)
print('''
scores: {} 
 - - - - - - - - - -
ㅣ score_mean: {:.2f} ㅣ
 - - - - - - - - - -
'''.format(scores, scores.mean()*100))
```

## Naive Bayse

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
scores = cross_val_score(clf, X, y, cv=kfold)
print('''
scores: {} 
 - - - - - - - - - -
ㅣ score_mean: {:.2f} ㅣ
 - - - - - - - - - -
'''.format(scores, scores.mean()*100))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
scores = cross_val_score(clf, X, y, cv=kfold)
print('''
scores: {} 
 - - - - - - - - - -
ㅣ score_mean: {:.2f} ㅣ
 - - - - - - - - - -
'''.format(scores, scores.mean()*100))
```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=30, learning_rate=1)
scores = cross_val_score(clf, X, y, cv=kfold)
print('''
scores: {} 
 - - - - - - - - - -
ㅣ score_mean: {:.2f} ㅣ
 - - - - - - - - - -
'''.format(scores, scores.mean()*100))
```
